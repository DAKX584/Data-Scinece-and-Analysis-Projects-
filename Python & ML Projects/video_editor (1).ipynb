### 

#### 4) yearly quaterly and monthly projects completed

```python
ve['Deadline'] = pd.to_datetime(ve['Deadline'], errors='coerce')
ve.dropna(subset=['Deadline'], inplace=True)

try:
    input_v = int(input('Please enter your desired year (e.g., 2024): '))
except ValueError:
    print('The entered value is not a valid year. Please try again.')

else:
    ve_f = ve[ve['Completion status'] == 'Completed'].copy()
    ve_f['year'] = ve_f['Deadline'].dt.year
    

    ve_year_specific = ve_f[ve_f['year'] == input_v]
    
   
    if ve_year_specific.empty:
        print(f'\nNo "Completed" task data is available for the year {input_v}.')

    else:
        ve_year_specific = ve_year_specific.copy()

        ve_year_specific['year'] = ve_year_specific['Deadline'].dt.year.astype(str)
        ve_year_specific['quarter'] = 'Q' + ve_year_specific['Deadline'].dt.quarter.astype(str)
        ve_year_specific['month'] = ve_year_specific['Deadline'].dt.month_name()
        
        yr = ve_year_specific.groupby('year')['Completion status'].count()
        Qut = ve_year_specific.groupby('quarter')['Completion status'].count()
        Mth = ve_year_specific.groupby('month')['Completion status'].count()
        
        
        print(yr.to_string())
        print('')
        print(Qut.to_string())
        print('')
        print(Mth.to_string())
```

#### 3) yearly monthly and quaterly pay

```python
ve['payment_date'] = pd.to_datetime(ve['payment_date'], errors='coerce')
ve.dropna(subset=['payment_date'], inplace=True)

try:
    year_input = int(input("Please enter the year you want to filter by (e.g., 2020): "))
except ValueError:
    print("Invalid input. Please enter a valid number for the year.")
else:
    ve_filtered = ve[ve['payment_date'].dt.year == year_input].copy()

    if ve_filtered.empty:
        print(f"No data found for the year {year_input}.")
    else:
        ve_filtered['year'] = ve_filtered['payment_date'].dt.year
        ve_filtered['quarter'] = 'Q' + ve_filtered['payment_date'].dt.quarter.astype(str)
        ve_filtered['month'] = ve_filtered['payment_date'].dt.month_name()
        
        
        Yr = ve_filtered.groupby('year')['amount'].sum()
        Qut =ve_filtered.groupby('quarter')['amount'].sum()
        Mth =ve_filtered.groupby('month')['amount'].sum()

        print(Yr.to_string())
        print('')
        print(Qut.to_string())
        print('')
        print(Mth.to_string())
```
